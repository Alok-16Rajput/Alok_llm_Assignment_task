# getting the relevant features

```python
columns_needed = {
    "eventTime": "_source.data.aws.eventTime",  # Timestamp of the event
    "userIdentity": "_source.data.aws.userIdentity.principalId",  # User who performed the action
    "eventName": "_source.data.aws.eventName",  # The action/event name
    "sourceIPAddress": "_source.data.aws.sourceIPAddress"  # Source IP address
}
```

# creating the new dataframe for handling missing values

```python
df_cleaned = df[list(columns_needed.values())].copy()
# Renaming the columns
df_cleaned.columns = list(columns_needed.keys())
```

```python
# Define a risk scoring function
def assign_risk_score(event_name, ip_address):
    risk_score = 0.0

    if "Delete" in str(event_name):
        risk_score = 0.8
    elif "Admin" in str(event_name):
        risk_score = 0.7
    elif ip_address and not ip_address.startswith("192.168"):  # Assuming external IPs are risky
        risk_score = 0.6

    return risk_score

# Apply risk scoring function
df_cleaned["risk_score"] = df_cleaned.apply(
    lambda row: assign_risk_score(row["eventName"], row["sourceIPAddress"]), axis=1
)

# Save to CSV
cleaned_file_path = "cleaned_data.csv"
df_cleaned.to_csv(cleaned_file_path, index=False)

```

## LLM implementation using Hugging Face

```python
from transformers import pipeline

model = pipeline("text-generation", model="google/flan-t5-base")

def analyze_logs(query, logs):
    prompt = f"""Security Log Analysis Task:
    Logs: {logs}
    Question: {query}
    Answer:"""
    return model(prompt, max_length=200)[0]['generated_text']

# Example usage (replace with your actual logs and query)
logs = df_cleaned.to_string() # Assuming df_cleaned is defined in your previous code
query = "What are the riskiest events?"
analysis_result = analyze_logs(query, logs)
analysis_result
```
